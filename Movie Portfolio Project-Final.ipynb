```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\User\Documents\Bibah DA\Correlation Python\movies.csv')

df.head()
```

```python
# missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for each columns

print(df.dtypes)
```

```python
# Change data type
df['budget'] = df['budget'].astype('Int64')
df['gross'] = df['gross'].astype('Int64')
```

```python
df[['budget','gross']].info()
df[['budget','gross']].head()
```

```python
df
```

```python
# duplicate value

df[df.duplicated()]
```

```python
df['yearcorrect'] = df['released'].str.extract(r'(\d{4})').astype('Int64')
df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
company_gross = (
    df.groupby('company')['gross']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)
print(company_gross)
company_gross.plot(kind='barh')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['budget'], bins=30)
plt.title('Distribution of Budget')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['gross'], bins=30)
plt.title('Distribution of Gross Revenue')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['gross'])
plt.title('Gross Outliers')
plt.show()
```

```python
# Top Genre
df['genre'].value_counts().head(10)
```

```python
plt.figure(figsize=(10,5))
df['genre'].value_counts().head(10).plot(kind='bar')
plt.title('Top Genres')
plt.show()
```

```python
df['country'].value_counts().head(10)
```

```python
plt.figure(figsize=(10,5))
df['country'].value_counts().head(10).plot(kind='bar')
plt.title('Top Countries')
plt.show()
```

```python
top10 = df.sort_values(by='gross', ascending=False)[
    ['name','year','gross']
].head(10)

sns.barplot(
    x='gross',
    y='name',
    data=top10
)
```

```python
df.sort_values(by='budget', ascending=False)[
    ['name','year','budget']
].head(10)
```

```python
yearly_gross = (
    df.groupby('yearcorrect')['gross']
      .sum()
      .reset_index()
)
print(yearly_gross)
plt.plot(yearly_gross['yearcorrect'],
         yearly_gross['gross'])
```

```python
df.sort_values(
    by='score',
    ascending=False
)[['name','score','yearcorrect']].head(10)
```

```python
top_score = (
    df[['name','score','year']]
      .sort_values('score',
                   ascending=False)
      .head(10)
)

sns.barplot(
    x='score',
    y='name',
    data=top_score
)
```
